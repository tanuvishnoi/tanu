```python
pca=PCA(n_components=11)
pca_df=pd.DataFrame(pca.fit_transform(ppp))
pca_df.head()
```

```python
from sklearn.cluster import KMeans
cluster_range = range( 1, 15 )
cluster_errors = []
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 10 )
  clusters.fit(pca_df)
  labels = clusters.labels_
  centroids = clusters.cluster_centers_
  cluster_errors.append( clusters.inertia_ )
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
clusters_df[0:15]
```

```python
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

```python
for i in range(len(clusters_df)-1):
    print(i+2,'-',i+1,': ',cluster_errors[i+1]-cluster_errors[i])
```

```python
kmeans = KMeans(n_clusters=3, n_init = 15, random_state=1)
prediction= kmeans.fit_predict(pca_df)
pca_df["group"] = prediction
ppp['group_pca']=prediction
```

```python
x=pca_df.drop('group',axis=1)
for i in x:
    print('relation between ',i,' and group: ')
    sns.boxplot(y=pca_df[i],x=pca_df['group'])
    plt.show()
```

```python
x=pca_df.drop(['group'],axis=1)
y=pca_df['group']
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.metrics import average_precision_score, classification_report
from sklearn.metrics import f1_score, precision_score, accuracy_score
# Function for evaluation metrics:
def report_cal(y_true, y_pred):
    class_report = classification_report(y_true, y_pred)
    accuracyscore = accuracy_score(y_true, y_pred)
    

    return  round(accuracyscore,3), print(class_report)
    


# Takes in a model, trains the model, and evaluates the model on the test set
def fit_and_evaluate(model):
    
    # Train the model
    model.fit(xtrain, ytrain)
    
    # Make predictions and evalute
    model_pred = model.predict(xtest)
    model_class_report = report_cal(ytest, model_pred)
    
    # Return the performance metric
    return model_class_report
```

```python
lr = LogisticRegression()
pca_lr_classreport = fit_and_evaluate(lr)

print('Accuracy Score for Logistic :', pca_lr_classreport)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
pca_knn_classreport = fit_and_evaluate(knn)

print('Accuracy Score for KNN :', pca_knn_classreport)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
pca_nb_classreport = fit_and_evaluate(nb)

print('Accuracy Score for Naive Bayes :', pca_nb_classreport)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
pca_dt_classreport = fit_and_evaluate(dt)

print('Accuracy Score for Decision Tree :', pca_dt_classreport)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf =RandomForestClassifier()
pca_rf_classreport = fit_and_evaluate(rf)

print('Accuracy Score for RandomForestClassifier :', pca_rf_classreport)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
pca_lr_classreport = fit_and_evaluate(lr)

print('Accuracy Score for LogisticRegression :', pca_lr_classreport)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb =GradientBoostingClassifier()
pca_gb_classreport = fit_and_evaluate(lr)

print('Accuracy Score for GradientBoosting :', pca_gb_classreport)
```
